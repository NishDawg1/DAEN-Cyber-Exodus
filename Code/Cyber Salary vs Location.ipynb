```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the CSV file
data = pd.read_csv('INDEED-GLASSDOOR-MERGED.csv')
```

```python
# 1. Location vs Lower Salary
lower_salary = data.groupby('Location')['Salary - lower bound'].min()
lower_salary.plot(kind='bar')
plt.title('Location vs Lower Salary')
plt.xlabel('Location')
plt.ylabel('Salary')
plt.show()
```

```python
# 2. Location vs Median Salary
median_salary = data.groupby('Location')['Salary - median'].median()
median_salary.plot(kind='bar')
plt.title('Location vs Median Salary')
plt.xlabel('Location')
plt.ylabel('Salary')
plt.show()
```

```python
# 3. Location vs Upper Salary
upper_salary = data.groupby('Location')['Salary - upper bound'].max()
upper_salary.plot(kind='bar')
plt.title('Location vs Upper Salary')
plt.xlabel('Location')
plt.ylabel('Salary')
plt.show()
```
